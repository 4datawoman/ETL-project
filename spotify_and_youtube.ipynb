# Spotify's Worldwide Daily Song Ranking
source: https://www.kaggle.com/edumucelli/spotifys-worldwide-daily-song-ranking

```python
daily_song_data = "Resources/daily_song_ranking.csv"

# daily_song_ranking.csv filter
song_ranking_df = pd.read_csv(daily_song_data)
song_ranking_df.head()
```

```python
US_song_ranking_df = song_ranking_df.drop(columns=['URL', 'Position', 'Date'])
US_song_ranking_df = US_song_ranking_df.loc[US_song_ranking_df["Region"] == "us"]
US_song_ranking_df = US_song_ranking_df.rename(columns={"Track Name": "Title"})
US_song_ranking_df.head()
```

```python
US_song_ranking_df.count()
```

```python
US_song_ranking_df = US_song_ranking_df.dropna(how='any')
```

```python
US_song_ranking_df.count()
```

```python
grouped_US_ranking_df = US_song_ranking_df.groupby(['Title','Artist'])['Streams'].sum().to_frame('Streams').reset_index()

grouped_US_ranking_df = grouped_US_ranking_df[['Artist', 'Title', 'Streams']]

sorted_US_ranking_df = grouped_US_ranking_df.sort_values(['Streams'], ascending=False)

sorted_US_ranking_df.head()

sorted_US_ranking_df.to_csv("US_daily_song_ranking.csv", index=False, header=True)
```

# Trending YouTube Videos
source: https://www.kaggle.com/datasnaek/youtube-new#USvideos.csv

```python
# read USvideos csv
videos_data = "Resources/youtube_videos.csv"

# youtube_videos.csv filter
videos_data_df = pd.read_csv(videos_data)
videos_data_df.head()
```

```python
musicvideos_data_df = videos_data_df.loc[videos_data_df["category_id"] == 10]
musicvideos_data_df.head()
```

```python
# trim df for certain columns
trimmed_musicvideos_df = musicvideos_data_df.loc[:, ['title', 'category_id', 'views']]

trimmed_musicvideos_df.category_id = trimmed_musicvideos_df.category_id.replace({10: "music"})

# ordered_musicvideos_df = trimmed_musicvideos_df.sort_values(['views'], ascending=False)
```

```python
grouped_musicvideos_df = trimmed_musicvideos_df.groupby(['title'])['views'].sum().to_frame('views').reset_index()

# sorted_musicvideos_df = grouped_musicvideos_df.sort_values(['views'], ascending=False)

# sorted_musicvideos_df.reset_index()

grouped_musicvideos_df.head()
```

```python
dropped_musicvideos_df = grouped_musicvideos_df[~grouped_musicvideos_df.title.str.contains('BTS')]
```

```python
dropped_musicvideos_df['title'].replace({'Official Video': ''}, inplace=True, regex=True)
dropped_musicvideos_df['title'].replace({'Official Music Video': ''}, inplace=True, regex=True)
dropped_musicvideos_df['title'].replace({'Official MV': ''}, inplace=True, regex=True)
dropped_musicvideos_df['title'].replace({'Lyric Video': ''}, inplace=True, regex=True)
dropped_musicvideos_df['title'].replace({'Official Audio': ''}, inplace=True, regex=True)
dropped_musicvideos_df['title'].replace({'Audio': ''}, inplace=True, regex=True)
dropped_musicvideos_df['title'].replace({'Music Video': ''}, inplace=True, regex=True)
dropped_musicvideos_df['title'].replace({'Vertical Video': ''}, inplace=True, regex=True)
dropped_musicvideos_df['title'].replace({'Official': ''}, inplace=True, regex=True)
dropped_musicvideos_df['title'].replace({'OFFICIAL MUSIC VIDEO': ''}, inplace=True, regex=True)
dropped_musicvideos_df['title'].replace({'Vertical Video': ''}, inplace=True, regex=True)
dropped_musicvideos_df['title'].replace({'Lyric Video': ''}, inplace=True, regex=True)
dropped_musicvideos_df['title'].replace({'OFFICIAL VIDEO': ''}, inplace=True, regex=True)
dropped_musicvideos_df['title'].replace({'Original Video': ''}, inplace=True, regex=True)


grouped_videos_df = dropped_musicvideos_df.groupby(['title'])['views'].sum().to_frame('views').reset_index()

grouped_videos_df = grouped_videos_df.sort_values(['views'], ascending=False)

grouped_videos_df.to_csv("Resources/music_videos.csv", index=False, header=True)
```
